Частина 1

1. Побудувати стовпчасту діаграму за допомогою бібліотеки Altair (Python).
Для цієї діаграми потрібно використовувати датасет source = data.wheat()
У діаграмі кожен стовпчик відповідає року, і його висота відображає кількість пшениці.
Якщо рік дорівнює 1810, то відповідний стовпчик буде виділено помаранчевим кольором, в інших випадках - синім.

```python
source = data.wheat()

chart = alt.Chart(source).mark_bar().encode(
    x='year:N',
    y='wheat:Q',
    color=alt.condition(
        alt.datum.year == 1810,
        alt.value('orange'),
        alt.value('blue')),
).properties(
    title='Стовпчаста діаграма кількості пшениці за роками'
)

chart
```

2. Побудувати стовпчасту діаграму допомогою бібліотеки Altair.
Для цього завдання вам потрібно використовувати датасет source = data.wheat().
Вимоги:
Створіть стовпчасту діаграму, де на осі X будуть роки, а на осі Y - кількість виробленої пшениці в мільйонах тонн.
Встановіть правило (horizontal rule) червоного кольору, яке показує середнє значення виробництва пшениці за всі роки.
Об'єднайте графік стовпчатої діаграми та правило в одному зображенні шириною 600 пікселів.
Забезпечте наявність легенди, що пояснює, як відображається середнє значення.
Збережіть графік у файлі з розширенням на вибір.

```python
source = data.wheat()

df = pd.DataFrame(source)

average_rule = alt.Chart(df).mark_rule(color='red').encode(
    y='mean(wheat):Q',
    size=alt.value(2))

bar_chart = alt.Chart(df).mark_bar().encode(
    x='year:O',
    y='wheat:Q',
    color=alt.condition(
        alt.datum.wheat > df['wheat'].mean(),
        alt.value('green'),
        alt.value('orange'))).properties(width=600)

final_chart = (bar_chart + average_rule).properties(
    title='Стовпчаста діаграма виробництва пшениці',
    height=300)

final_chart.save('wheat_production_chart.html')
final_chart
```

3. Побудувати  інтерактивний графік, використовуючи бібліотеку Altair.
Використовувати датасет source = data.stocks()
Тип графіка лінійний, з маркерами.
Вимоги:
На осі X буде відображено часові мітки (дата та час), на осі Y - ціни акцій,
а кожна лінія буде представляти окрему компанію.
Визначте кожну лінію графіка за допомогою кольорів
Додайте точки на графіку, які відображають конкретні значення цін акцій у визначені моменти часу.
Додайте легенду, що пояснює, яка лінія відповідає кожній компанії.
Збережіть графік у файл.

```python
source = data.stocks()

chart = alt.Chart(source).mark_line(point=True).encode(
    x='date:T',
    y='price:Q',
    color='symbol:N',
    tooltip=['symbol:N', 'date:T', 'price:Q']
).properties(
    width=600,
    height=400
).interactive()

chart.save('interactive_chart.html')
chart
```

1. Побудуйте звичайну лінійну діаграму на основі  pd.DataFrame з двома стовпцями
за допомогою бібліотеки Altair для функції y=x cos⁡x для значень x в діапазоні від 0 до 99.
Додайте відповідні мітки для вісей x та y на діаграмі. Підпишіть вісі і назву графіку.

```python
x = np.arange(0, 100, 1)
y = x * np.cos(x)
dt = pd.DataFrame({'x': x, 'y': y})

chart = alt.Chart(dt).mark_line().encode(
    x='x:Q',
    y='y:Q'
).properties(
    title='Графік функції y = x * cos(x)',
    width=600,
    height=400
).configure_axis(
    titleFontSize=12,
    labelFontSize=10
).configure_title(
    fontSize=14
)

chart
```

3. Побудуйте гістограму із від'ємними значеннями на основі даних
про зміни в зайнятості в США (data.us_employment()) за допомогою бібліотеки Altair.
Вісь x відповідає місяцям, вісь y відображає зміни в зайнятості (nonfarm_change).
Використовуйте два різних кольори для відображення від'ємних та додатних значень змін в зайнятості.
Встановіть ширину діаграми на 500 пікселів.

```python
employment_data = data.us_employment()

chart = alt.Chart(employment_data).mark_bar().encode(
    x=alt.X('month:N', title='Місяць'),
    y=alt.Y('nonfarm_change:Q', title='Зміни в зайнятості'),
    color=alt.condition(
        alt.datum.nonfarm_change > 0,
        alt.value('steelblue'),
        alt.value('orange')
    )
).properties(
    width=500,
    title='Зміни в зайнятості в США'
)

chart
```

4.	Побудуйте гістограму з групуванням на основі даних про врожай ячменю (data.barley()).
Вісь x представляє роки, вісь y відображає сумарний врожай (yield), кольорова шкала вказує
на різні роки, а групування (стовпці) відбувається за різними місцями (site).

```python
barley_data = data.barley()

chart = alt.Chart(barley_data).mark_bar().encode(
    x=alt.X('year:O', title='Рік'),
    y=alt.Y('sum(yield):Q', title='Сумарний врожай'),
    color=alt.Color('year:N', title='Рік'),
    column=alt.Column('site:N', title='Місце')
).properties(
    width=150,
    title='Гістограма врожаю ячменю з групуванням за роками та місцями'
)

chart
```

чисте виробництво електроенергії (net_generation), а кожен шар діаграми

7.	Створіть діаграму розсіювання для двовимірної гістограми на основі даних
про фільми (data.movies.url). Вісь X відображає рейтинг фільмів на IMDB (IMDB_Rating),
вісь Y відображає рейтинг фільмів на Rotten Tomatoes (Rotten_Tomatoes_Rating).

```python
movies_data = data.movies.url

scatter_plot = alt.Chart(movies_data).mark_circle(size=60).encode(
    x=alt.X('IMDB_Rating:Q', bin=alt.Bin(maxbins=30), title='IMDB Rating'),
    y=alt.Y('Rotten_Tomatoes_Rating:Q', bin=alt.Bin(maxbins=30), title='Rotten Tomatoes Rating'),
    color='count()',
).properties(
    width=600,
    height=400,
    title='Діаграма розсіювання для двовимірної гістограми фільмів'
)

scatter_plot
```
